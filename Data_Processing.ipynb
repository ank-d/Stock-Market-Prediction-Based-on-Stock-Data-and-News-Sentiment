```python
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IndianFinancialNews.csv')
data
```

```python
start_date='2017-02-27'
end_date='2020-02-25'
data['Date'] =pd.to_datetime(data['Date'],infer_datetime_format=True)


data = data[ (data['Date'] >= start_date) & (data['Date'] <= end_date) ]
data = data.sort_values(by='Date', ascending=True)
data=data[['Date','Title','Description']]
data = data.reset_index(drop=False)
data=data[['Date','Title','Description']]
data
data=data.head()


```

```python
from transformers import AutoModelForSequenceClassification
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoConfig
from scipy.special import softmax
import tensorflow as tf
import torch

```

```python
# MODEL = "mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis"
MODEL = "ProsusAI/finbert"

# MODEL = "nickmuchi/finbert-tone-finetuned-finance-topic-classification"

tokenizer = AutoTokenizer.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL)

#
# TF
# model = TFAutoModelForSequenceClassification.from_pretrained(MODEL)
model = TFAutoModelForSequenceClassification.from_pretrained(MODEL)
model.save_pretrained(MODEL)
tokenizer.save_pretrained(MODEL)
```

```python
def polarity(text):
    encoded_input = tokenizer(text, return_tensors='tf')
    output = model(encoded_input)
    print(1)
    # logits = output.logits
    # sc = torch.softmax(logits, dim=-1)
    scores = output[0][0].numpy()
    # abc=output[0][1].numpy()
    scores = softmax(scores)

    return (scores[0],scores[1],scores[2],)
```

```python
tweets=data
tweets['Negative'],tweets['Neutral'],tweets['Positive'] = zip(*tweets['Description'].apply(lambda txt:polarity(txt) ))
tweets.to_csv("polarizedNews.csv",index=False)
```

```python
new=pd.read_csv("polarizedNews.csv")

def calculate_sentiment(row):
  max_sentiment = max(row['Negative'], row['Positive'], row['Neutral'])
  if max_sentiment == row['Negative']:
    return -1
  elif max_sentiment == row['Positive']:
    return 1
  else:
    return 0

new['Polarity'] = new.apply(calculate_sentiment, axis=1)


Pol_df = pd.DataFrame(new.groupby('Date')['Polarity'].sum())
Pol_df.to_csv("MeanSentiment.csv",index=True)

Pol_df=Pol_df.reset_index()
Pol_df['Date']=pd.to_datetime(Pol_df['Date']).dt.date
Pol_df=pd.DataFrame(Pol_df.set_index('Date',drop=True))
# new.head()

nifty_df = pd.read_csv("NIFTY.csv")


# Extract the date part
nifty_df['Date'] = pd.to_datetime(nifty_df['Date']).dt.date
# nifty_df
nifty_df=nifty_df.set_index('Date',drop=True)
nifty_df=pd.DataFrame(nifty_df)
final_df = nifty_df.join(Pol_df,how="inner")
# final_df = pd.merge(nifty_df,Pol_df,on='Date')


final_df.to_csv('FinalMeta_Data.csv')
# Pol_df.head()
# Pol_df.info()
# nifty_df.head()
# nifty_df.info()
```
